```python
from pathlib import Path
import shutil

scratch_folder = Path("/scratch/radv/gemosquerarojas/")

selected_predictions_folder = scratch_folder / "selected_predictions_GS"
patients_data = Path("/scratch/radv/share/glioseg/new_run_corrected/Patients/")

selected_predictions_folder.mkdir(exist_ok=True, parents=True)


for patient in sorted(patients_data.iterdir()):  

    segmentations_dir = patient / "SEGMENTATIONS" / "NIFTI"
    segmentation_best_gs = segmentations_dir / "mask_tumor_nnunet_task500_ensemble_relabelled_T1GD.nii.gz"
    segmentation_gs_staple = segmentations_dir / "mask_tumor_ensemble_T1GD.nii.gz"

    print(segmentation_best_gs)
    print(segmentation_gs_staple)

    patient_dest = selected_predictions_folder / patient.name
    patient_dest.mkdir(exist_ok=True, parents=True)

    shutil.copy(segmentation_best_gs, patient_dest / "best_gs.nii.gz")
    shutil.copy(segmentation_gs_staple, patient_dest / "gs_staple.nii.gz")
```
